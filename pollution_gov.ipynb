## detecting outliers

```python
from scipy import stats

def detect_outliers(column):
    z_scores = np.abs(stats.zscore(column))
    return z_scores > 3

outliers = df[detect_outliers(df['pollutant_min'])]
print(outliers)
```

# Now EDA

## *Distribution of Pollutant Values*

```python
plt.figure(figsize=(10,6))
plt.hist(df['pollutant_min'], bins=50, alpha=0.5, label='Min')
plt.hist(df['pollutant_max'], bins=50, alpha=0.5, label='Max')
plt.hist(df['pollutant_avg'], bins=50, alpha=0.5, label='Avg')
plt.legend()
plt.title('Distribution of Pollutant Values')
plt.xlabel('Pollutant Value')
plt.ylabel('Frequency')
plt.show()
```

## *Top 10 Cities by Average Pollutant Value*

```python
top_cities = df.groupby('city')['pollutant_avg'].mean().sort_values(ascending=False).head(10)
print(top_cities)
```

## *Top 10 Pollutants by Average Value*

```python
top_pollutants = df.groupby('pollutant_id')['pollutant_avg'].mean().sort_values(ascending=False).head(10)
print(top_pollutants)
```

## *Correlation Matrix*

```python
corr_matrix = df[['pollutant_min', 'pollutant_max', 'pollutant_avg']].corr()
print(corr_matrix)
```

# *Bar Chart: Top 10 Cities by Average Pollutant Value*

```python

top_cities = df.groupby('city')['pollutant_avg'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(15,8))
plt.bar(top_cities.index, top_cities.values)
plt.xlabel('City')
plt.ylabel('Average Pollutant Value')
plt.title('Top 10 Cities by Average Pollutant Value')
plt.show()
```

# *Map: Average Pollutant Value by State*

```python
import plotly.express as px

state_avg = df.groupby('state')['pollutant_avg'].mean().reset_index()

fig = px.choropleth(state_avg, locations="state", locationmode="USA-states", color="pollutant_avg",
                    scope="usa",
                    color_continuous_scale="Viridis",
                    range_color=(0, 100),
                    labels={'pollutant_avg':'Average Pollutant Value'})

fig.update_layout(title_text='Average Pollutant Value by State', title_x=0.5)
fig.show()
```

# *Bar graph: Total Pollutant Value by State*

```python
state_total = df.groupby('state')['pollutant_avg'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(15,6))
plt.bar(state_total.index, state_total.values)
plt.xlabel('State')
plt.ylabel('Total Pollutant Value')
plt.title('Total Pollutant Value by State')
plt.show()
```

# *"Top 10 Pollution Peaks" Area Chart*

```python
import matplotlib.pyplot as plt

state_max = df.loc[df.groupby('state')['pollutant_max'].idxmax().head(10)]

plt.figure(figsize=(18,6))
plt.fill_between(state_max['state'], state_max['pollutant_max'], alpha=0.5)
plt.xlabel('State')
plt.ylabel('Maximum Pollutant Value')
plt.title('Pollution Peaks: Maximum Pollutant Values by State')
plt.show()
```

# *"Top States of Pollution" Radar Chart*

```python
import matplotlib.pyplot as plt

state_max = df.loc[df.groupby('state')['pollutant_max'].idxmax().head()]

plt.figure(figsize=(10,6))
plt.plot(state_max['state'], state_max['pollutant_max'], 'o-')
plt.xlabel('State')
plt.ylabel('Maximum Pollutant Value')
plt.title('State of Pollution: Maximum Pollutant Values by State')
plt.show()
```

# *Pollution Heatmap: Maximum Pollutant Values by City and State*

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,8))
sns.heatmap(df.pivot_table(index='city', columns='state', values='pollutant_max'), annot=True, cmap='Reds')
plt.title('Pollution Heatmap: Maximum Pollutant Values by City and State')
plt.xlabel('State')
plt.ylabel('City')
plt.show()
```

# *"Pollution by Latitude" Scatter Plot*

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.scatter(df['latitude'], df['pollutant_max'])
plt.xlabel('Latitude')
plt.ylabel('Maximum Pollutant Value')
plt.title('Pollution by Latitude: Maximum Pollutant Values')
plt.show()
```

# *Pollution by Top 10 Stations: Distribution of Maximum Pollutant Values*

```python
import matplotlib.pyplot as plt

# Get the top 10 stations with the highest maximum pollutant values
top_stations = df.groupby('station')['pollutant_max'].max().sort_values(ascending=False).head(10)

# Create a box plot for the top 10 stations
plt.figure(figsize=(15,6))
plt.boxplot([df[df['station'] == station]['pollutant_max'] for station in top_stations.index])
plt.xlabel('Station')
plt.ylabel('Maximum Pollutant Value')
plt.title('Pollution by Top 10 Stations: Distribution of Maximum Pollutant Values')
plt.xticks(range(1, 11), top_stations.index, rotation=90)  # Rotate the x-axis labels for better readability
plt.show()
```

# *"Pollutant Levels by Latitude and Longitude" Scatter Plot*

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.scatter(df['latitude'], df['longitude'], c=df['pollutant_max'])
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.title('Pollutant Levels by Latitude and Longitude')
plt.colorbar(label='Pollutant Max')
plt.show()
```

# *"City-wise Pollutant Levels by Pollutant Type" Bar Chart*

```python
import matplotlib.pyplot as plt

city_pollutants = df.groupby(['city', 'pollutant_id'])['pollutant_avg'].mean().unstack().head()

plt.figure(figsize=(15,6))
city_pollutants.plot(kind='bar')
plt.xlabel('City')
plt.ylabel('Average Pollutant Value')
plt.title('City-wise Pollutant Levels by Pollutant Type')
plt.show()
```
